```python
%matplotlib inline
import numpy as np 

from astropy.table import Table 
from astropy.coordinates import SkyCoord
import astropy.units as u
import matplotlib.pyplot as plt
import matplotlib as mpl 
mpl.rcParams['font.family'] = 'stixgeneral'

from matplotlib.patches import Circle
from matplotlib.collections import PatchCollection
```

Read in stellar mass, impact para, and projected X/Y distances

```python
# dwarf qso pairs that we use in the paper
plt_tb = Table.read('data/zheng_dwarfcgm-survey_cut-mstar9.5-snr8-br200m.csv', format='ascii')
logMstar_sub = plt_tb['logM*']
logMstar_sub_err = plt_tb['logM*_err']

gal_qso_dist = plt_tb['impact_para_kpc']
gal_qso_dist_err = plt_tb['impact_para_err']

gal_qso_rvir = plt_tb['R200m_kpc']
gal_qso_rvir_err = plt_tb['R200m_err']

rho_norm_sub = np.around(gal_qso_dist/gal_qso_rvir, decimals=3)

aa = gal_qso_dist_err/gal_qso_dist
bb = gal_qso_rvir_err/gal_qso_rvir
rho_norm_sub_err = np.around(rho_norm_sub*np.sqrt((aa)**2 + (bb)**2), decimals=3)
```

Calculate projected X/Y distances 

```python
gal_coord = SkyCoord(ra=plt_tb['gal_ra_deg'], dec=plt_tb['gal_dec_deg'], 
                     frame='icrs', unit=(u.deg, u.deg))
qso_coord = SkyCoord(ra=plt_tb['qso_ra_deg'], dec=plt_tb['qso_dec_deg'], 
                     frame='icrs', unit=(u.deg, u.deg))

del_ra = (qso_coord.icrs.ra.deg-gal_coord.icrs.ra.deg)
del_dec = (qso_coord.icrs.dec.deg-gal_coord.icrs.dec.deg)
del_sep_deg = np.sqrt(del_ra**2 + del_dec**2)
#del_sep_deg = dw_coord.separation(qso_incoord).arcmin
del_sep_kpc = gal_qso_dist
del_sep_kpc_err = gal_qso_dist_err

# use negative in xkpc because ra is large on the left 
# so left is East 
del_xkpc = -np.around((del_ra/del_sep_deg)*del_sep_kpc, decimals=1)
del_xkpc_err = np.abs(np.around((del_ra/del_sep_deg)*del_sep_kpc_err, decimals=1))

del_ykpc = np.around((del_dec/del_sep_deg)*del_sep_kpc, decimals=1)
del_ykpc_err = np.abs(np.around((del_dec/del_sep_deg)*del_sep_kpc_err, decimals=1))

del_xkpc_norm = del_xkpc / gal_qso_rvir
del_xkpc_norm_err = np.abs(del_xkpc_norm*np.sqrt((del_xkpc_err/del_xkpc)**2 + (gal_qso_rvir_err/gal_qso_rvir)**2))

del_ykpc_norm = del_ykpc / gal_qso_rvir
del_ykpc_norm_err = np.abs(del_ykpc_norm*np.sqrt((del_ykpc_err/del_ykpc)**2 + (gal_qso_rvir_err/gal_qso_rvir)**2))
```

Some plotting setups 

```python
refs = ['Bordoloi+2014', 'Johnson+2017', 
         'Liang&Chen2014', 'Qu&Bregman2022', 'Zheng+2019b', 
          'Zheng+2020b', 'NewObs/Archived']
symbols = ['s', 'D', 'p', 'X', '*', '<', 'o']
ss_err = [8, 8, 10, 10, 12, 10, 8]

impact_label = r'Impact Parameter $b$ (kpc)/ $R_{\rm 200m}$'
mstar_label = r'log$_{\rm 10}$(M$_*$/M$_\odot$)'

fs = 16
```

```python
# we plot detections for CIV 
ion = 'CIV'
flg = plt_tb['logN_flg_'+ion]
ind_det = flg == '='
ind_nondet = np.logical_not(ind_det)
```

```python
cur_ref = 'New Pairs Added\nin This Work'
dot_color = plt.cm.Greys(0.5)
edge_color = 'm'
```

```python
logmstar_median = 8.3  # median stellar mass 
impact_median = 0.45 # median impact parameter 
```

```python
fig = plt.figure(figsize=(15, 5))

ax1 = fig.add_axes([0.06, 0.23, 0.25, 0.75]) # logMstar vs b plot
ax2 = fig.add_axes([0.39,0.23, 0.25, 0.75], aspect=1) # normalized xkpc vs ykpc 
ax3 = fig.add_axes([0.725, 0.23, 0.26, 0.75], aspect=1) # x vs y kpc in phy units 

###### panel 1 #### 
# let's not plot error bar here since only nearby stuff have good uncertainty estimates
y = logMstar_sub
yerr = logMstar_sub_err + np.nan
x = rho_norm_sub
xerr = rho_norm_sub_err + np.nan

for i, iref in enumerate(refs): 
    ind_ref = plt_tb['References'] == iref
    if iref == 'Current': 
        iref = cur_ref

    ax1.errorbar(x[ind_ref & ind_nondet], y[ind_ref & ind_nondet], 
                xerr=xerr[ind_ref& ind_nondet], 
                yerr=yerr[ind_ref& ind_nondet], fmt=symbols[i], 
                markersize=ss_err[i], 
                markerfacecolor=dot_color, 
                markeredgewidth=0.5, 
                markeredgecolor='k', ecolor=dot_color, label=iref)
    
    ax1.errorbar(x[ind_ref & ind_det], y[ind_ref & ind_det], 
                xerr=xerr[ind_ref& ind_det], 
                yerr=yerr[ind_ref& ind_det], fmt=symbols[i], 
                markersize=ss_err[i], 
                markerfacecolor=dot_color, 
                markeredgewidth=2, 
                markeredgecolor=edge_color, ecolor=dot_color, label=None)

ax1.axvline(impact_median, color='k', ls=':', lw=2)
ax1.axhline(logmstar_median, color='k', ls=':', lw=2)
ax1.tick_params(labelsize=fs)
ax1.minorticks_on()
ax1.set_xlabel(impact_label, fontsize=fs+2)
ax1.set_ylabel(mstar_label, fontsize=fs+2); 


##### panel 2 ### 
ax2.scatter([0], [0], marker='2', s=80, color='k', lw=2, label='host dwarfs')
for i, iref in enumerate(refs): 
    ind_ref = plt_tb['References'] == iref
    if 'Zheng+2019' in iref:
        continue
    if iref == 'ThisWork': 
        iref = cur_ref

    ax2.errorbar(del_xkpc_norm[ind_ref & ind_nondet], del_ykpc_norm[ind_ref & ind_nondet], 
                xerr=del_xkpc_norm_err[ind_ref & ind_nondet], 
                yerr=del_ykpc_norm_err[ind_ref & ind_nondet], fmt=symbols[i], 
                markersize=ss_err[i]-0.5, 
                markerfacecolor=dot_color, 
                markeredgewidth=0.5, 
                markeredgecolor='k', ecolor='k', label=iref)
    ax2.errorbar(del_xkpc_norm[ind_ref & ind_det], del_ykpc_norm[ind_ref & ind_det], 
                xerr=del_xkpc_norm_err[ind_ref & ind_det], 
                yerr=del_ykpc_norm_err[ind_ref & ind_det], fmt=symbols[i], 
                markersize=ss_err[i]-0.5, 
                markerfacecolor=dot_color, 
                markeredgewidth=1.5, 
                markeredgecolor=edge_color, ecolor=dot_color, label=None)
    
#### add circles 
c50 = Circle((0, 0), 0.5)
c100 = Circle((0, 0), 1.)
patches = [c50, c100]
p = PatchCollection(patches, alpha=0.4, facecolor='none', edgecolor='k', ls='--')
ax2.add_collection(p) 
ax2.tick_params(axis='both', labelsize=fs)
ax2.text(0.18, -0.55, r'0.5R$_{\rm 200m}$', fontsize=fs-4, alpha=0.7, rotation=40)
ax2.text(0.52, -0.92, r'R$_{\rm 200m}$', fontsize=fs-4, alpha=0.7, rotation=40)
ax2.set_xlabel(r'X (kpc) / $R_{\rm 200m}$', fontsize=fs)
ax2.set_ylabel(r'Y (kpc) / $R_{\rm 200m}$', fontsize=fs)
ax2.set_yticks([-1, -0.5, 0, 0.5, 1])
ax2.set_xlim(-1.1, 1.1)
ax2.set_ylim(-1.1, 1.1)
ax2.minorticks_on()

# add arrows to indicate north and east 
ax2.arrow(1, 0.8, -0.1, 0., color=plt.cm.Greys(0.7), head_width=0.03, lw=1, head_length=0.03)
ax2.arrow(1, 0.8, 0, 0.1, color=plt.cm.Greys(0.7), head_width=0.03, lw=1, head_length=0.03)
ax2.text(0.97, 0.95, 'N', color=plt.cm.Greys(0.7), fontsize=12)
ax2.text(0.79, 0.78, 'E', color=plt.cm.Greys(0.7), fontsize=12)

##### panel 3 #### 
ax3.scatter([0], [0], marker='2', s=80, color='k', lw=2, label='host dwarfs')
for i, iref in enumerate(refs): 
    ind_ref = plt_tb['References'] == iref
    if iref == 'Current': 
        iref = cur_ref
    ax3.errorbar(del_xkpc[ind_ref & ind_nondet], del_ykpc[ind_ref & ind_nondet], 
                xerr=del_xkpc_err[ind_ref & ind_nondet], 
                yerr=del_ykpc_err[ind_ref & ind_nondet], fmt=symbols[i], 
                markersize=ss_err[i]-0.5, 
                markerfacecolor=dot_color, 
                markeredgewidth=0.5, 
                markeredgecolor='k', ecolor=dot_color, label=iref)
    
    ax3.errorbar(del_xkpc[ind_ref & ind_det], del_ykpc[ind_ref & ind_det], 
                xerr=del_xkpc_err[ind_ref & ind_det], 
                yerr=del_ykpc_err[ind_ref & ind_det], fmt=symbols[i], 
                markersize=ss_err[i]-0.5, 
                markerfacecolor=dot_color, 
                markeredgewidth=1.5, 
                markeredgecolor=edge_color, ecolor=dot_color, label=None)
    
c50 = Circle((0, 0), 50)
c100 = Circle((0, 0), 100)
c150 = Circle((0, 0), 150)
patches = [c50, c100, c150]
p = PatchCollection(patches, alpha=0.4, facecolor='none', edgecolor='k', ls='--')
ax3.add_collection(p)
    
ax3.tick_params(axis='both', labelsize=fs)
ax3.text(18, -55, '50 kpc', fontsize=fs-5, alpha=0.7, rotation=40)
ax3.text(50, -95, '100 kpc', fontsize=fs-5, alpha=0.7, rotation=40)
ax3.text(75, -140, '150 kpc', fontsize=fs-5, alpha=0.7, rotation=40)
ax3.set_xlabel('X (kpc)', fontsize=fs)
ax3.set_ylabel('Y (kpc)', fontsize=fs)
ax3.minorticks_on()

# add arrows to indicate north and east 
ax3.arrow(150, 115, -15, 0., color=plt.cm.Greys(0.7), head_width=5, lw=1, head_length=5)
ax3.arrow(150, 115, 0, 15, color=plt.cm.Greys(0.7), head_width=5, lw=1, head_length=5)
ax3.text(145, 138, 'N', color=plt.cm.Greys(0.7), fontsize=12)
ax3.text(120, 110, 'E', color=plt.cm.Greys(0.7), fontsize=12)

# add legend box 
ax4 = fig.add_axes([0.02, 0.02, 0.965, 0.07]) # legend box 
ax4.set_xticks([])
ax4.set_yticks([])
ax4.spines['bottom'].set_color(plt.cm.Greys(0.5))
ax4.spines['top'].set_color(plt.cm.Greys(0.5)) 
ax4.spines['right'].set_color(plt.cm.Greys(0.5))
ax4.spines['left'].set_color(plt.cm.Greys(0.5))
ax4.set_xlim(0, 1)
ax4.set_ylim(0, 1)

dx = 0.1
ax4.scatter([0.013], [0.45], marker='2', s=80, color='k', lw=2)
ax4.text(0.022, 0.3, 'Host Dwarfs', fontsize=fs-3)
# Bordoloi
ax4.scatter([0.013+dx], [0.45], marker=symbols[0], s=80, color=plt.cm.Greys(0.6))
ax4.text(0.022+dx, 0.3, refs[0], fontsize=fs-3)
# Johnson
ax4.scatter([0.013+dx*2+0.005], [0.45], marker=symbols[1], s=80, color=plt.cm.Greys(0.6))
ax4.text(0.022+dx*2+0.005, 0.3, refs[1], fontsize=fs-3)
# LC14
ax4.scatter([0.013+dx*3+0.01], [0.45], marker=symbols[2], s=80, color=plt.cm.Greys(0.6))
ax4.text(0.022+dx*3+0.01, 0.3, refs[2], fontsize=fs-3)
# QB22
ax4.scatter([0.013+dx*4+0.035], [0.45], marker=symbols[3], s=80, color=plt.cm.Greys(0.6))
ax4.text(0.022+dx*4+0.035, 0.3, refs[3], fontsize=fs-3)
# Zheng20
ax4.scatter([0.013+dx*5+0.06], [0.45], marker=symbols[5], s=80, color=plt.cm.Greys(0.6))
ax4.text(0.022+dx*5+0.06, 0.3, refs[5][:-1], fontsize=fs-3)
# new obs or ar
ax4.scatter([0.013+dx*6+0.06], [0.45], marker=symbols[-1], s=80, color=plt.cm.Greys(0.6))
ax4.text(0.022+dx*6+.06, 0.3, 'New obs or archival data', fontsize=fs-3)
ax4.text(0.022+dx*8+0.01, 0.3, 'Magenta Symbols = Detections', color='m',fontsize=fs-3)
fig.savefig('figs/fig3_logMstar_vs_b_XYkpc.pdf'); 
```
